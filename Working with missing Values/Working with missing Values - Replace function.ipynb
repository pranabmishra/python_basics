# Working with missing Values - Replace function

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('weather_data.csv.txt')
```

```python
df
```

### How to replace values with replace function?

```python
new_df = df.replace(-99999, np.NaN)

# to replace values with NaN 
```

```python
new_df
```

```python
new_df2 = df.replace([-99999,31.0], np.NaN)

# to replace Two Values in a DF. have to supply all data with a list
```

```python
new_df2
```

### Replace values based on specific column

```python
df
```

```python
df.replace({
    'temperature': -99999,
    'windspeed' : -99999,
    'event' : '0'
},np.NaN)
```

```python
df.replace({ 
    -99999 : np.NaN,
    '0': 'Sunny'   
})
```
