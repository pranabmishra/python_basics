# How to sort a Series or DataFrame?

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.title.sort_values().head()
```

```python
movies['title'].sort_values().head()
```

#### sort by particular column

```python
movies.sort_values('duration').head()
```

```python
movies.sort_values('duration', ascending = False).head()
```

```python
movies.sort_values('star_rating').head()
```

```python
movies.sort_values(['star_rating', 'duration'],ascending = False).head()
```
