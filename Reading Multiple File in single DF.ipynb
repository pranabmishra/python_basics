```python
files = [file for file in os.listdir('./new_data')]

all_files = pd.DataFrame()

for file in files:
    df = pd.read_csv('./new_data/'+file)
    all_files = pd.concat([all_files,df])
    
all_files

# adding all files to a sinh]gle data set to Explore
```
