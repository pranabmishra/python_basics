## How to plot multiple data in a single plot?

```python
days = [1,2,3,4,5,6,7]
max_t = [45,50,53,56,67,69,76]
min_t = [34,37,43,51,55,57,61]
avg_t = [40,35,45,51,57,67,75]
```

```python
plt.figure(figsize = (10,5)) # To change the size of the graph


plt.plot(days,min_t, label = 'min_t')
plt.plot(days,avg_t, label = 'Avg_t')

plt.xlabel('Days', size = 15)
plt.ylabel('Temperature', size = 15)

plt.title('Weather Data', size = 15)

plt.legend(loc = 'upper left', fontsize = 'large')

plt.grid()
plt.show()
```

```python
company = ['Google','Amazon', 'MSFT', 'FB']
revenue = [90,136, 89, 27]
```

```python
plt.bar(company, revenue, label = "Revenue")

plt.legend()

plt.xlabel('Company Name')
plt.ylabel('Revenue in Price')
plt.show()
```

```python
company = ['Google','Amazon', 'MSFT', 'FB']
revenue = [90,136, 89, 27]
profit = [40,10,34,12]
```

```python
plt.bar(company, revenue, label = "Revenue")
plt.bar(company,profit, label= 'Profit')

plt.legend()

plt.xlabel('Company Name')
plt.ylabel('Revenue in Price')
plt.show()
```

```python
plt.figure(figsize = (10,6))

plt.bar(company, revenue, label = "Revenue")
plt.plot(company,profit, label= 'Profit', color = 'red')

plt.legend()

plt.title(' Company Profit Graph', size = 15)
plt.xlabel('Company Name')
plt.ylabel('Revenue in Price')
plt.show()
```

```python
xpos = np.arange(len(company))
```

```python
plt.figure(figsize = (10,6))

plt.bar(xpos- 0.2, revenue, label = "Revenue")
plt.bar(xpos,profit, label= 'Profit', color = 'yellow')


plt.xticks(xpos,company)

plt.legend()

plt.title(' Company Profit Graph', size = 15)
plt.xlabel('Company Name')
plt.ylabel('Revenue in Price')
plt.show()
```

```python
plt.figure(figsize = (10,6))

plt.bar(xpos- 0.2,revenue, width = 0.3, label = "Revenue")
plt.bar(xpos,profit,width = 0.4, label= 'Profit', color = 'pink')


plt.xticks(xpos,company)

plt.legend()

plt.title(' Company Profit Graph', size = 15)
plt.xlabel('Company Name')
plt.ylabel('Revenue in Price')
plt.show()
```

```python
plt.figure(figsize = (10,6))

plt.barh(xpos- 0.2, revenue, label = "Revenue")
plt.barh(xpos,profit, label= 'Profit', color = 'yellow')


plt.yticks(xpos,company)

plt.legend()

plt.title(' Company Profit Graph', size = 15)
plt.xlabel('Company Name')
plt.ylabel('Revenue in Price')
plt.show()
```
