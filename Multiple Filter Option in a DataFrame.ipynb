# How to Apply multiple Filter Option in  a DataFrame?

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[movies.duration >= 200]

# . notation can be used only when all column names without space inbetween words
```

```python
movies[movies['duration'] >= 200]

# call the entire column when any space is with column name
```

### Filterring with logical operator

#### AND Condition

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]

# use the () to give direct of valuation method and use & instead of and 
```

```python
movies[(movies['duration'] >= 200) & (movies['genre'] == 'Drama')]

# 2nd Method 
```

#### OR condition

```python
movies[(movies.duration >= 200) | (movies.genre == 'Drama')].head()
```

```python
movies[movies.genre.isin(['Crime','Drama', 'Adventure'])].head()

# for a specific row value
```
