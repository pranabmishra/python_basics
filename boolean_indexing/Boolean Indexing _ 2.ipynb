### Write a function find_missing that has 3 parameters , df, col1 and col2 where df is teh data frameand col1 and col2 are col names. This function Should return  all the rows of DataFrame where col1 and col2 are missing.

```python
# write a function find_missing values in 2 column 

def find_missing(df, col1, col2):
    filt1 = df[col1].isna()
    filt2 = df[col2].isna()
    filt = filt1 & filt2
    cols = [col1,col2]
    return df.loc[filt,cols]
```

```python
find_missing(movie,'gross','budget').head()
```
